# Problemas de valor inicial

## $ \S 1 $ Introdução

Uma equação envolvendo uma função e suas derivadas com respeito a uma ou mais variáveis independentes é chamada de **equação diferencial**. Muitas das leis e fenômenos da Física, Astronomia, Química e Biologia são expressas naturalmente através de equações diferenciais. Como exemplos, o decaimento radioativo, as leis da Mecânica e de movimentos oscilatórios, a equação de taxa de reações químicas e o crescimento de populações biológicas são todos governados, ou ao menos bem modelados, por equações diferenciais. De fato, como a derivada de uma função representa sua taxa de variação, por exemplo com respeito ao tempo ou a uma coordenada espacial, é freqüente que exista uma relação quantitativa entre esta taxa de variação e a própria função; geralmente esta relação toma a forma de uma equação diferencial.

## $ \S 2 $ Exemplo

De acordo com a segunda lei de Newton, a aceleração escalar $ a $ experimentada por um objeto  é proporcional à magnitude total $ F $ da força que age sobre ela, com uma constante de proporcionalidade $ \frac{1}{m} $, onde $ m $ é a massa do objeto. Em símbolos:
$$
a = \frac{1}{m} F \quad \text{ou equivalentemente} \quad F = ma.
$$
Por definição, a aceleração é a taxa de variação com respeito ao tempo $ t $ da velocidade, que por sua vez é a taxa de variação da posição. 

Suponha que o objeto em questão esteja em queda livre, sob ação exclusiva da gravidade, e seja $ y $ a sua distância de uma determinada altura. Então vale a equação diferencial
$$
F = mg = m\frac{d^2y}{dt^2},
$$
ou
$$
\frac{d^2y}{dt^2} = g,
$$
onde $ g $ é a aceleração devido à gravidade.

Se agora levarmos em consideração que durante a queda o ar imprime uma força de resistência proporcional à velocidade do objeto, a segunda lei de Newton implica que seu movimento é modelado pela equação diferencial
$$
F = mg - k\frac{dy}{dt} = m \frac{d^2y}{dt^2} \qquad (k \text{ constante}).
$$

📝 Recorde as notações alternativas:
$$ \frac{dy}{dt} = y' = \dot y ,\quad  \frac{d^2y}{dt^2} =y'' = \ddot y,\quad \frac{d^3y}{dt^3} =y''' = y^{(3)} = \dddot y \quad \cdots \quad \frac{d^ny}{dt^n} = y^{(n)} = \overset{\,n}{\dot{y}}.
$$
Assim, a última equação diferencial também poderia ter sido escrita de forma mais concisa como:
$$
m y'' = mg - ky' \quad \text{ou} \quad m\ddot y = mg - k \dot y \qquad (k \text{ constante}).
$$

## $ \S 3 $ Taxonomia de equações diferenciais

### $ 3.1 $ Equações diferenciais ordinárias e parciais

Uma **equação diferencial ordinária (EDO)** é uma equação que envolve derivadas de uma função de uma *única* variável independente. A função (variável dependente) freqüentemente é denotada por $ y $, enquanto a variável independente geralmente é denotada por $ t $ ou $ x $. Por exemplo:
* (a) $ \displaystyle{\frac{dy}{dt} = cy} $, onde $ c $ é constante;
* (b) $ \displaystyle{m\ddot y = -ky} $, onde $ k $ é constante;
* (c) $ \displaystyle{y' +2xy = x^2e^{-x}} $;
* (d) $ \displaystyle{\ddot y - 3 \dot y + 2y = 0} $;
* (e) $ \displaystyle{\frac{d^2y}{dt^2} + (\alpha + \beta \cos t) y = 0} $, onde $ \alpha,\,\beta $ são constantes;
* (f) $ \displaystyle{x^3y^{(3)} +x^2y'' − 2 x y' +2 y = 2x^4} $;
* (g) $ \displaystyle{\frac{d^{2}y}{dx^{2}}+k\frac{dy}{dx}+m\left(\frac {dy}{dx}\right)^{3}+n^{2}y=0} $, onde $k,\,m,\,n $ são constantes;
* (h) $ \displaystyle{y''=\frac{1}{\sqrt {t}}y^{3/2}} $.

Uma **equação diferencial parcial (EDP)** é uma equação envolvendo as derivadas *parciais* de uma função de *duas ou mais* variáveis independentes. A função (variável dependente) freqüentemente é denotada por $ u $. Alguns exemplos são:
* $ \displaystyle{u_t + ku_x} = 0 $, onde $ k $ é constante;
* $ \displaystyle{\frac{\partial^2u}{\partial x^2} + \frac{\partial^2u}{\partial y^2}  = 0} $ (*equação de Laplace em duas dimensões*);
* $ \displaystyle{u_{xx} + u_{yy} + u_{zz} = u_{t}} $ (*equação do calor em três dimensões*);
* $ \displaystyle{\frac{\partial^2u}{\partial t^2} = c^2\bigg(\frac{\partial^2u}{\partial x_1^2} + \frac{\partial^2u}{\partial x_2^2} + \cdots + \frac{\partial^2u}{\partial x_n^2} \bigg)}$, onde $ c > 0 $ é constante (*equação da onda em $ n $ dimensões*).

### $ 3.2 $ A ordem de uma equação diferencial

Dizemos que uma equação diferencial é **de ordem $ r $** ($ r \ge 1 $) se ela envolve a (ou uma) derivada de ordem $ r $ da função, mas não as derivadas de ordem $ > r $.

Os exemplos (a) e (c) da lista de EDOs acima são de primeira ordem; os exemplos (b), (d), (e), (g) e (h) são de segunda ordem; e o exemplo (f) é uma equação diferencial ordinária de terceira ordem. As equações diferenciais parciais mencionadas são todas de segunda ordem, exceto pela primeira da lista, que é de primeira ordem.

📝 A maioria das equações diferenciais que modelam fenômenos naturais é de ordem $ 1 $ ou $ 2 $.

## $ \S 4 $ Forma normal de uma EDO

A equação diferencial ordinária de ordem $ r $ mais geral tem a forma
$$
g\big(t, y, y', \dots, y^{(r)}\big) = 0.
$$
Contudo, assumiremos que uma equação deste tipo pode ser reescrita na **forma normal**
$$
y^{(r)} = f\big(t, y, y', \dots, y^{(r-1)}\big).

## $ \S 5 $ Soluções de equações diferenciais ordinárias e condições iniciais

### $ 5.1 $ Problemas de valor inicial de primeira ordem

Uma função $ \varphi(t) $ definida num intervalo é uma **solução** da EDO de primeira ordem 
$$
y' = f(t, y)
$$
se ela satisfaz
$$
\varphi'(t) = f(t, \varphi(t))
$$
para todo $ t $ em seu domínio.

Assim como ocorre com o conjunto de primitivas de uma função dada, há infinitas soluções para uma equação deste tipo, que envolvem uma constante arbitrária, a chamada **constante de integração**. Para determinar uma solução particular, precisamos conhecer o valor dela em algum ponto, i.e., é preciso que se especifique o valor de $ y $ num determinado $ t $, digamos:
$$
y(t_0) = y_0.
$$
Esta última é chamada **condição inicial**.

### $ 5.2 $ Problemas de valor inicial de ordem $ r $

Considere uma equação diferencial ordinária de ordem $ r $:
\begin{equation*}\label{E:ode}
y^{(r)} = f\big(t, y, y', \dots, y^{(r-1)}\big)\ . \tag{1}
\end{equation*}

De modo geral, a determinação de uma solução única requer o conhecimento de $ r $ condições auxiliares. Se estas condições especificam o valor de $ y $ e suas $ r - 1 $ primeiras derivadas num mesmo ponto $ t_0 $:
\begin{equation*}\label{E:initial}
y(t_0) = y_0, \quad y'(t_0) = y_0', \quad y''(t_0) = y_0'' \quad \dots \quad y^{(r-1)}(t_0) = y_0^{(r-1)}\ , \tag{2}
\end{equation*}
então elas são chamadas de **condições iniciais**. Observe que apesar da notação, os $ y_0^{(k)} $ são *números*, e não derivadas de uma função $ y_0 $.

Uma equação diferencial ordinária \eqref{E:ode} juntamente com um conjunto de condições iniciais de mesma ordem \eqref{E:initial} formam um **problema de valor inicial (PVI)**. Uma **solução** deste PVI é uma função $ \varphi $ que é solução de \eqref{E:ode} e satisfaz
\begin{equation*}
\varphi(t_0) = y_0, \quad \varphi'(t_0) = y_0', \quad \varphi''(t_0) = y_0'' \quad \dots \quad \varphi^{(r-1)}(t_0) = y_0^{(r-1)}\ .
\end{equation*}

📝 Sob certas condições razoáveis sobre $ f $, o *Teorema de existência e unicidade de soluções para EDOs* garante que existe $ \varepsilon > 0 $ tal que o problema de valor inicial \eqref{E:ode}$ + $\eqref{E:initial} possui uma única solução no intervalo $ [t_0 - \varepsilon, t_0 + \varepsilon] $. Entretanto muitas vezes esta solução não pode ser estendida a uma função definida em toda a reta $ \mathbb R $.

### $ \S 6 $ Equações diferenciais ordinárias de ordem $ r $ como sistemas de $ r $ equações diferenciais de ordem $ 1 $

Introduzindo as funções auxiliares
$$
y_0 = y, \quad y_1 = y', \quad y_2 = y'' \quad \dots \quad y_{r-1} = y^{(r-1)}
$$
transformamos a EDO
$$
y^{(r)} = f\big(t, y, y', \dots, y^{(r-1)}\big)
$$
num *sistema* de $ r $ equações de primeira ordem:
$$
\begin{cases}
y_0' = y_1 \\ 
y_1' = y_2 \\ 
\ \vdots \\
y_{r-2}' = y_{r-1} \\
y_{r-1}' = f(t,y_0,y_1,\dots,y_{r-1})
\end{cases}
$$
ou equivalentemente, usando a notação vetorial,
$$
\boldsymbol y = (y_0,y_1,\dots,y_{r-1})\,,
$$
temos
$$
\boldsymbol y' =
\begin{bmatrix}
y_0' \\
y_1' \\
\vdots \\
y_{r-2}' \\
y_{r-1}'
\end{bmatrix}=
\begin{bmatrix}
y_1\vphantom{y_0'} \\
y_2\vphantom{y_1'} \\
\vdots \\
y_{r-1}\vphantom{y_{r-2}'} \\
f(t,\boldsymbol y)\vphantom{y_{r-1}'}
\end{bmatrix}
= \boldsymbol F(t,\boldsymbol y).
$$
Nesta notação, as condições iniciais \eqref{E:initial} podem ser escritas como
$$
\boldsymbol y(t_0) = \boldsymbol y_0 = \big(y_0, y_0', y_0'', \dots, y_0^{(r-1)}\big).
$$
Portanto, *qualquer problema de valor inicial de ordem $ r $ é equivalente a um sistema de $ r $ equações diferenciais de ordem $ 1 $ e $ r $ condições iniciais, ou seja, a um PVI vetorial de ordem $ 1 $ em $ \mathbb R^r $*, da forma:
\begin{equation*}\label{E:vetorial}
\begin{cases}
\boldsymbol y' = \boldsymbol F(t, \boldsymbol y) \\
\boldsymbol y(t_0) = \boldsymbol y_0
\end{cases}
\quad \text{onde} \quad \boldsymbol y = \big(y_0, y_1, \dots, y_{r-1}\big).\tag{3}
\end{equation*}

## $ \S 7 $ Soluções numéricas de problemas de valor inicial

### $ 7.1 $ Soluções analíticas e soluções numéricas

Em geral as soluções **analíticas** (i.e., exatas) de uma equação diferencial não podem ser expressas através de fórmulas fechadas. Isto ocorre mesmo para a equação diferencial mais simples, em que $ f $ não dependente de $ y $:
$$
y' = f(t)\,,
$$
cuja solução é equivalente à determinação das primitivas de $ f $.

Ademais, mesmo quando é possível descrever as soluções explicitamente, isto pode ser muito trabalhoso. Por estes motivos é comum o uso de *métodos numéricos* para solução de equações diferenciais.

Uma **solução numérica** de um problema de valor inicial
$$
\begin{cases}
\boldsymbol y' = \boldsymbol F(t, \boldsymbol y) \\
\boldsymbol y(t_0) = \boldsymbol y_0
\end{cases}
$$
é simplesmente uma tabela de valores $ t_i $ e $ \boldsymbol y_i $ ($ i= 0, 1, \dots, N$), com $ \boldsymbol y_i $ o mais próximo possível do valor preciso $ \boldsymbol y(t_i) $ da solução analítica.

### $ 7.2 $ Determinação de soluções numéricas

Nos métodos que estudaremos, o procedimento para obtenção de uma solução numérica é sempre o seguinte. Começamos escolhendo o  **tamanho do passo** $ h \ne 0 $ e então tomamos os **nodos** 
$$
t_i = t_0 + ih \qquad (i = 0, 1, \dots, N)
$$
como $ N + 1 $ pontos igualmente espaçados dentro do intervalo $ [t_0, t_0 + Nh] $ (ou do intervalo $ [t_0 + Nh, t_0] $, caso $ h < 0 $). Finalmente, partindo do valor inicial $ \boldsymbol y_0 $ dado, de alguma maneira (que varia com o método) calculamos a aproximação $ \boldsymbol y_i $ para $ \boldsymbol y(t_i) $ usando a aproximação $ \boldsymbol y_{i-1} $ anterior, para $ i = 1, 2, \dots, N $ sucessivamente.

📝 Além de tabelar os valores obtidos numericamente, sempre que possível é recomendado representá-los graficamente com auxílio de um computador.
